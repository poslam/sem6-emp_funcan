```python
import sympy as sp
import matplotlib.pyplot as plt
```

```python
n, t = sp.symbols("n t")

f = t**2 / sp.exp(2 * n * t)

integ = n**3 * sp.integrate(f, (t, 0, 2))

integ
```

```python
f2 = sp.Abs(sp.exp(-n / (1 - t**2)))**2

integ2 = sp.integrate(f2, (t, -1, 1))

integ2.simplify()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def x_n_a(t, n):
    return (t * n * np.sqrt(n)) / np.exp(n)


def x_n_b(t, n):
    result = np.exp(-n / (1 - t**2))
    result[np.abs(t) >= 1] = 0  # Условие для |t| >= 1
    return result


def test(t, n):
    return np.exp(-2 * n * t)


# Границы для графиков
t_a = np.linspace(0, 2, 100)
t_b = np.linspace(-2, 2, 200)

# Выбираем несколько значений n
n_values = [1, 2, 5, 10, 20]

# --- Вариант (a) ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for n in n_values:
    plt.plot(t_a, x_n_a(t_a, n), label=f"n={n}")

plt.axhline(0, color="black", linestyle="--", label="x(t)=0")
plt.xlabel("t")
plt.ylabel("x_n(t)")
plt.title("Равномерная сходимость x_n(t) для 1")
plt.legend()

# --- Вариант (b) ---
plt.subplot(1, 2, 2)
for n in n_values:
    plt.plot(t_b, x_n_b(t_b, n), label=f"n={n}")

plt.axhline(0, color="black", linestyle="--", label="x(t)=0")
plt.xlabel("t")
plt.ylabel("x_n(t)")
plt.title("Равномерная сходимость x_n(t) для 2")
plt.legend()

plt.tight_layout()
plt.show()
```
